```python
import numpy as np
```

NUMPY INDEXING AND SLICING OF VECTORS

```python
arr = np.arange(11,21)
arr
```

```python
arr[0]  # indexing starts from zero
arr[1]
arr[2]
arr[3]
arr[4]
arr[5]
arr[6]
arr[7]
arr[8]
arr[9]
```

```python
# slicing
# arr[start : end(excluded)]
arr_slice = arr[2:5]  #  index 2 to 4
arr_slice
```

```python
# arr[ : end]  --> start(0) to 6
arr_slice = arr[:6]
arr_slice
```

```python
# arr[start : ]   -->> start(3) to end
arr_slice = arr[3 : ]
arr_slice
```

```python
# arr[ : ]   -->>> start to end all element
arr_slice = arr[ : ]  # complete row
arr_slice
```

```python
# arr[start : end : step(gap)]   ---->> start(2) to end(6) with gap 2 --> 13,15,17
arr_slice = arr[2 : 8 : 2]
arr_slice
```

```python
arr_slice = arr[1 : : 2]
arr_slice
```

NUMPY INDEXING AND SLICING OF MATRIX

```python
arr = np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0]   # print 0th row
arr[1]   # print 1th row etc.
```

```python
 #  single element finding
#  arr[row_index, column_index]
arr[0,0]  
```

```python
#  fnding a part of matrix
#  arr[row_start : row_end(excluded) , column_start : column_end(excluded)]
part_slice = arr[0:2 , 1:4]
part_slice
```

```python
part_slice1 = arr[3:6 , 3:5]
part_slice1
```

```python
part_slice2 = arr[3:,3:]   # row goes 3 to last and column goes 3 to last
part_slice2
```

```python
# print columns
#  arr[:,column_index]  --->> select all rows & given column
arr[:,4]  # print column 3
```

BOOLEAN INDEXING

```python
arr = np.arange(1,11)
arr
```

```python
bool_index = arr % 2 == 0
bool_index
```

```python
arr = arr[bool_index]  # in array where value are true they will kept and remove others false values
arr
```
