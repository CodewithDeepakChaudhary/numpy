2. Array generation function

```python
arr1 = np.arange(1,11)  # this generate 1 to 10 countinuously. last no not included.
arr1
```

```python
arr2 = np.arange(1,11,2)  # this generate between 1 to 11 with the gap of 2. last no not included.
arr2
```

```python
arr3 = np.zeros(6)  # generate zeros according passing values. here 6
arr3
```

```python
arr4 = np.zeros((5,6))   # generate zeros matrix according passing values. here 5 x 6
arr4
```

```python
arr5 = np.ones(5)  # this generate ones. 5 times.
arr5
```

```python
arr6 = np.ones((5,6))  # this generate ones matrix . 5x6
arr6
```

```python
arr7 = np.linspace(1,11,6)  # this gives a linear space (equal space between all nubmers) between given range. last parameter tells how many numbers you want.
arr7
```

3. Random Generation function

```python
np.random.rand(10)   # randomaly generates numbers according input. This is also called Normlization (all values between (0 - 1)).
```

```python
np.random.randn(10)  # this is also generates random numbers according input. This is also called Standirization (all values between (-3 - 3)).
```

```python
np.random.randint(5)  # this generate only one integer number bettween 0 to 5.
```

```python
np.random.randint(2,10,5)  # this generate  5 numbers between 2 to 10 . last parameter tells how many no. you want to generate.
```

6. Array Methods

```python
arr
```

```python
arr.min()  # give minimum value 
```

```python
arr.max() # give maximum value
```

```python
arr.sum()  # sum all of values in matrix
```

```python
np.sum(arr, axis = 0)  # sum of values column wise (axis = 0)
```

```python
arr.mean()  # give mean of matrix of all numbers.
```

```python
arr.std()  # give standard deviation of matrix of all numbers.
```

```python
arr.argmax()  # gives the index of maximum value of a matrix.
```
