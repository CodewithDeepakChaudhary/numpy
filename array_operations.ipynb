ARRAY OPERATIONS

Arithmetic operations   ---->  for arithmetic operations both array size must be same.

```python
a1 = np.array([1,2,3,4,5])
a2 = np.array([6,7,8,9,10])
```

```python
# for arithmetic operations both array size must be same
a1 + a2
```

```python
a1 - a2
```

```python
a1 * a2
```

```python
a1 / a2
```

```python
a1 // a2  # floor division
```

```python
a1 ** a2  # power
```

Broadcasting

```python
list = [10,20,30,40]
arr= np.array(list)
```

```python
for i in list:
    print(i+10)
```

```python
arr + 10  #  broadcasting using array , which is not possible in list . (above method follow in list)
```

Deep copy and Shallow copy

```python
# shallow copy --> in shallow copy when we slice a vectoe or matrix, changes  made will be reflected  nested in original vector or matrix.
# deep copy ---> create new copy of array. not affected original array.
a = np.arange(1,21)
a
```

```python
slice = a[:5]
slice = slice * 10
slice
```

```python
a   # not changes in original array
```

```python
a[4:6] *10
```

```python
a
```

```python
# b = a.copy()  # here if we changed in b , if .copy used then original array not changed
b = a   # here if we changed in b then original array also changed
b
```

```python
b[0] = 99
b
```

ADVANCE ARRAY MANIPULATION

```python
# we can not stack two matrix. because they already stacked. so,
# we can stack two arrays(vectors)

a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

```

```python
a
```

```python
b
```

```python
np.vstack((a,b))  # vertically stack add row wise
```

```python
np.hstack((a,b))  # horizontally stack join two arrays
```

```python
np.column_stack((a,b))  # column stack make column of arrays
```

```python
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)  ## split array in to 2 equl 2D array  horizontaly
```

```python
np.hsplit(c,4)  # split array in to 4 equl 2D array horizontaly
```

```python
np.vsplit(c,2)  ## split array in to 2 equl 2D array vertically
```

```python
a = np.vsplit(c,4)   ### split array in to 4 equl 2D array vertically
a
```

```python
for i in a:
    print(i)
```
